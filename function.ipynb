## 1. What is the difference between a function and a method in Python?
- Function
    - A function is a block of code that is defined using the def keyword (or lambda) and is not tied to any object or class.

    - It can be called independently and used to perform a specific task.
  
Example:
    
     def add(a, b):
    return a + b

    print(add(3, 5))
    Output: 8
       here add is a function because it's defined globally and can be used without being associated with any object.

- Method
    - A method is a function that is associated with an object.

    - It is called on an object and can access and modify the object’s internal state (if it's an instance method).

    - Methods are defined inside a class.

Example:

    text = "hello world"
    print(text.upper())  
    Output: HELLO WORLD
      upper() is a method because it is called on a string object (text), and it performs an action using that object’s data.


## 2. Explain the concept of function arguments and parameters in Python.

- A function in Python is a block of reusable code that performs a specific task when called.

Example:

    def greet(name):
    print(f"Hello, {name}!")
- Parameters

Definition:
Parameters are placeholders defined in the function definition.
They act as variables that receive values when the function is called.

Where: Inside the parentheses during function definition.

Example:

    def greet(name):  # ← "name" is a parameter
    print(f"Hello, {name}!")
- Arguments

Definition:
Arguments are the actual values passed to the function when it is called.

Where: Inside the parentheses during function call.

Example:

    greet("Alice")  # ← "Alice" is an argument

## 4. What is the purpose of the `return` statement in a Python function?
- The return statement in Python is used to end the execution of a function and send a value back to the caller. It defines what output the function should produce when it is called.If no return is used, the function returns None by default.The returned value can be stored in a variable, printed, or used in expressions.

Example:

    def multiply(x, y):
    return x * y  # returns the product of x and y

    result = multiply(3, 4)
    print(result)          Output: 12
Here, return x * y sends the result (12) back to the caller.
The value is stored in the variable result and then printed.

Without return, the function only prints the value but does not return it, so result is None.



## 5. What are iterators in Python and how do they differ from iterables?
- Iterable
An iterable is any Python object capable of returning its members one at a time using a loop (like for).

It implements:

    _iter_()method


 Examples: Lists, tuples, strings, dictionaries, sets.
- Iterator
An iterator is an object that keeps state and knows how to fetch the next value when requested using next().

It implements:

    __iter__() → returns itself

    __next__() → returns next item or raises StopIteration

Example:

     Iterable :-
    my_list = [1, 2, 3]
    for item in my_list:        # Can be iterated
    print(item)

     Convert iterable to iterator :-
    my_iter = iter(my_list)     # Now it's an iterator
    print(next(my_iter))         Output: 1
    print(next(my_iter))         Output: 2

## 6. Explain the concept of generators in Python and how they are defined.
- A generator is a special type of iterator that produces values one at a time using the yield keyword instead of returning them all at once.
Generators are memory-efficient because they don't store all values in memory but generate them on demand.

Key Points:

Memory-efficient: Generates values on the fly (lazy evaluation).

Stateful: Remembers its execution state between calls.

Defined using def and yield, not return.

Automatically implements __iter__() and __next__().

  - Defining a Generator:

examlple


    def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1    Output: 1  2   3

This function returns a generator object, not a final result.

Each call to next() will resume from the last yield.



## 7. What are the advantages of using generators over regular functions?
- Advantages:

  - Efficient Memory Usage: Generators do not store the entire sequence in memory, making them suitable for large datasets or infinite sequences.

  - Simpler Syntax: Using yield simplifies the creation of iterators, making the code more concise and readable.

  - Lazy Evaluation:
    - On-Demand Data: Generators produce values one at a time and only when required, avoiding the need to generate or store all values upfront.

   - Memory Efficiency: Since values are generated as needed and not all at once, generators use minimal memory.

 - Maintainable Code:
Code using yield is often more readable and cleaner than managing lists and loops manually.

  - Infinite Sequences:
Generators can represent sequences without limits

  - State Retention:
Generators pause at each yield, resume where they left off — unlike functions that restart on each call.

example:

    def count_up_to(n):
    for i in range(1, n+1):
        yield i

    gen = count_up_to(3)

    for num in gen:
    print(num)          output : 1  2  3

## 9. Explain the purpose and usage of the `map()` function in Python.
- Purpose:
The map() function in Python is used to apply a function to every item in an iterable (like a list, tuple, etc.) and return a new map object (which is iterable) containing the results.

-  Usaage: Simplifies code when you need to perform the same operation on every element in a sequence.

Syntax:

    
    map(function, iterable)

Example:


    nums = [1, 2, 3]
    doubled = map(lambda x: x * 2, nums)
    print(list(doubled))    Output: [2, 4, 6]



## 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?


These are built-in functional programming tools in Python used for processing iterables.

  - Map()
Purpose: Applies a function to each item in an iterable.
Returns: A new iterable with transformed values.

 Example:


    map(lambda x: x * 2, [1, 2, 3])  ➞  [2, 4, 6]

  - Filter():
   Purpose: Filters items in an iterable based on a condition (function returns True/False).

  Returns: A new iterable with filtered values.

 Example:

    
    filter(lambda x: x % 2 == 0, [1, 2, 3, 4])  ➞  [2, 4]

  - Reduce() (from functools module)
Purpose: Applies a function cumulatively to the items in an iterable to reduce it to a single value.
Returns: A single result.

 Example:


    from functools import reduce
    reduce(lambda x, y: x + y, [1, 2, 3, 4])  ➞  10

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2 or n > 200:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is a prime number.")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a

fib = FibonacciIterator(10)
for num in fib:
    print(num, end=" ")
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

for power in powers_of_two(5):
    print(power, end=" ")
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_lines(filename):
    with open(filename, 'r') as f:
        for line in f:
            yield line.strip()

 # Createing a sample file
with open('myfile.txt', 'w') as f:
    f.write("Hello\nWorld\nPython")


for line in read_lines('myfile.txt'):
    print(line)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(text):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, text))

input_str = "Hello, World!"
result = remove_vowels(input_str)
print("String without vowels:", result)
```
